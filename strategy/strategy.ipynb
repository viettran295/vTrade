```python
from vtrade import vTrade
from strategy import Strategy
# import plotly.io as pio
# pio.renderers.default = 'notebook_connected'

SHORT_MA = "EWM_20"
LONG_MA = "EWM_50"
vtr = vTrade()
df = vtr.get_stock_data("COIN")
```

```python
strategy = Strategy()
df = strategy.calc_MA(df, SHORT_MA)
df = strategy.calc_MA(df, LONG_MA)
```

```python
df_ma = strategy.calc_crossing_MA(df, SHORT_MA, LONG_MA)
strategy.show_crossing_MA(df_ma, SHORT_MA, LONG_MA)
```

```python
df_rsi = strategy.calc_RSI(df)
strategy.show_RSI(df_rsi)
```

```python
df_bb = strategy.calc_bollinger_bands(df)
strategy.show_bollinger_bands(df_bb)
```
